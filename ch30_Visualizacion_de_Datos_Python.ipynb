### 1. Comparar Datos

![bar-chart](https://github.com/efviodo/data-science/raw/master/courses/utec/figures/1_3SdKohVkzfDd0E0Mn_a54Q.png)

Para comparar datos, por ejemplo de diferentes variables de un mismo dataset, o datos de un dataset contra otro, los siguientes tipos de gráficos son muy útiles:

- [Barras](https://www.anychart.com/chartopedia/chart-type/bar-chart/): Permite comparar de forma directa, valores cuantitativos por variable o categoría
- [Pilas apliadas](https://www.anychart.com/chartopedia/chart-type/percent-stacked-bar-chart/): Permiten analizar valores cuantitativos por categoría y a la vez compararlos con diferentes muestras. Por ejemplo analizar el PBI de diferentes paises y teniendo en cuenta el aporte porcentual de cada sector.
- [Gráfico Radar](https://www.anychart.com/chartopedia/chart-type/radar-chart/): Es otra alternativa para comparar los valores cuantitativos de diferentes variables. Por ejemplo, mostrar el puntaje en diferentes habilidades que hace a la destreza de un jugador de futobl, como *Conversión de Goles*, *Precisión en pases*, etc.

<a id='Explorar-Datos'></a>
### 2. Explorar relaciones Part-To-Whole (de la parte al todo)

![pie-chart](https://github.com/efviodo/data-science/raw/master/courses/utec/figures/1_FaYgHfMCxO_pHIPxSeH6fQ.png)

Cuando lo que se quiere es explorar la composición por ejemplo de una variable categórica y analizar relaciones del tipo part-to-whole, los siguientes tipos de visualziaciones son muy útiles:

- [Gráfico de Torta](https://www.anychart.com/chartopedia/chart-type/pie-chart/) (Pie chart/Donut chart): Permite analizar de forma básica de la composición porcentual de una variable categórica.
- [Gráfico Piramidal](https://www.anychart.com/chartopedia/chart-type/pyramid-chart/) (Pyramid chart): Cuando los datos representan o responden a una "jerarquía", puede ser interesante utilizar este tipo de gráficos.
- [Mapa de Árbol](https://www.anychart.com/chartopedia/chart-type/treemap/) (Treemap chart): Permite representar relaciones jerárquicas complejas. Por ejemplo la cantidad de ingresos anuales de una compañía agregados por unidades de negocio y para cada unidad de negocio por proyecto.
- [Gráfico de Embudo](https://www.anychart.com/chartopedia/chart-type/funnel-chart/) (Funnel chart): Este tipo de gráfico se utiliza mucho para comparar variables cuantitativas, que de alguna forma estan relacionadas entre si como parte de un proceso. Por ejemplo, se puede utilizar para analizar ventas de un sitio de eComerce, ayudando a identificar los cueyos de botella del proceso.

<a id='Analisis-Temporal'></a>
### 3. Analisis en el tiempo

![time-charts](https://github.com/efviodo/data-science/raw/master/courses/utec/figures/1_cr1_zR9ruLixMdYbAZzrNQ.png)

Cuando se busca analizar la relación entre una o más variables con una componente temporal, como por ejemplo puede ser analizar la evolución de los salarios de Uruguay, respecto del tiempo, los siguientes tipos de visualizaciones son los indicados:

- [Gráfico de linea/curva](https://www.anychart.com/chartopedia/chart-type/line-chart/) (Line/Spline chart): Es la representación gráfica más simple y utilizada para mostrar la evolución de una variable cuantitativa en función de otra variable.
- [Gráfico de área]() (Área chart): A diferencia de la anterios, utiliza áreas para representar una variable cuantitativa.
- [Gráfico candela Japonesa](https://www.anychart.com/chartopedia/chart-type/japanese-candlestick-chart/) (Candlestick chart): Es muy útil cuando se queire analizar una variable que puede tomar múltiples valores para un mismo momento, por ejemplo analizar la evolución del precio de las acciones de una compañia frente a otras, tomando como referencia el precio de apertura y cierre de las acciones en la bolsa por día. Es una variante de las gráficas [OHLC](https://www.anychart.com/chartopedia/chart-type/ohlc-chart/).
- [Gráfico Sparkline](https://www.anychart.com/chartopedia/chart-type/sparkline-chart/) (Sparkline chart): Consiste en construír versiones simplificadas y pequeñas de otros gráficos temporales como Line Charts o Área Charts, eliminando por ejemplo los ejes del gráfico. Son muy útiles para utilizar en tablas, en donde para cada fila queremos hacer más visual una variable determinada. Por ejemplo listar los montos totales de exportaciones de granos en Uruguay en un año entero, utilizando Sparklines para mostrar para cada grano la variación mensual del precio de este.

<a id='Analisis-Distribucion'></a>
### 4. Analizar distribución de un dataset

![distribution](https://github.com/efviodo/data-science/raw/master/courses/utec/figures/1_rQkdb2kL3MfcWdtqMtf24w.png)

Para analizar la distribución de una variable en un determinado dataset y comparar por ejemplo con otros datasets, encontrar relaciones y co-relaciones, los siguientes tipo de gráficos son muy útles:

- [Histogramas](https://en.wikipedia.org/wiki/Histogram): Un histograma es una representación de la estimación de la distribución de valores de un dataset numérico.
- [Gráfico de dispersión de puntos](https://www.anychart.com/chartopedia/chart-type/dot-chart/) (Dot (scatter) chart): Permite examinar las tendencias en distribución y correlación entre dos variables.
- [Gráfico de burbujas](https://www.anychart.com/chartopedia/chart-type/bubble-chart/) (Bubble charts): Cada punto se grafica como un disco, expresando dos de los atributos/dimensiones del dato a través de la ubicación del disco en el sistema de ejes cartesianos y el tercero a través de su tamaño.
- [Gráfico Box-and-whisker](https://www.anychart.com/chartopedia/chart-type/box-chart/)
- [Gráfico de errores](https://www.anychart.com/chartopedia/chart-type/error-chart/)
- [Mapa de calor](https://www.anychart.com/chartopedia/chart-type/heatmap/) (Heatmap chart): Es una representación visual bidimensional de los datos, donde los valores se codifican en colores.
- [Gráfico de rangos](https://www.anychart.com/products/anychart/gallery/Range_Charts/) (Range chart)
- [Gráfico Polar](https://www.anychart.com/products/anychart/gallery/Polar_Charts/) (Polar chart)

### 6. Datos geográficos

![map](https://github.com/efviodo/idatha-data-science-course/blob/master/notebooks/figures/1_Iy0yitXT4mxejyGIX5EwOQ.png?raw=1)

Muchas veces, la mejor forma de darle sentido a datos geográficos (geo-referenciados) es utilizando visualziaciones en un mapa.

- [Mapa de Colores](https://www.anychart.com/chartopedia/chart-type/choropleth-map/) (Choropleth maps): Permiten identificar diferencias entre áreas geográficas.
- [Mapa de Puntos](https://www.anychart.com/chartopedia/chart-type/dot-map/) (Dot maps): Proporcionan una visualización accesible de la distribución geográfica de los fenómenos en las áreas geogŕaficas en la que se encuentran.
- [Mapa de Burbujas](https://www.anychart.com/chartopedia/chart-type/bubble-map/) (Bubble maps): Similar a los gráficos de burbujas, permiten agregar una tercera dimensión a la posición de un fenómeno, utilizando el tamaño de una burbujar. Son muy útiles para agregar la información de un mapa de puntos cuando contamos con tantos puntos que se hace difiícil entender la visualización.

- [Mapa de Conexiones](https://www.anychart.com/chartopedia/chart-type/connector-map/) (Connector map): Permite representar relaciones entre los distintos puntos del mapa mediante lineas que unen los puntos.

- [Mapa de Flujos](https://www.anychart.com/chartopedia/chart-type/flow-map/) (Flow map): Útiles para explorar cómo los objetos se mueven entre ubicaciones, en especial cuando interesa mostrar la dirección del flujo.

```python
import seaborn as sns
sns.set(style="darkgrid")

# Load an example dataset with long-form data
fmri = sns.load_dataset("fmri")

# Plot the responses for different events and regions
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="dark")
rs = np.random.RandomState(50)

# Set up the matplotlib figure
f, axes = plt.subplots(3, 3, figsize=(9, 9), sharex=True, sharey=True)

# Rotate the starting point around the cubehelix hue circle
for ax, s in zip(axes.flat, np.linspace(0, 3, 10)):

    # Create a cubehelix colormap to use with kdeplot
    cmap = sns.cubehelix_palette(start=s, light=1, as_cmap=True)

    # Generate and plot a random bivariate dataset
    x, y = rs.randn(2, 50)
    sns.kdeplot(x, y, cmap=cmap, shade=True, cut=5, ax=ax)
    ax.set(xlim=(-3, 3), ylim=(-3, 3))

f.tight_layout()
```

### Film Dialogue

Film dialogue es un proyecto de HANAH ANDERSON y MATT DANIELS, quienes buscaron estudiar que tan homogenea es la participación de actores femeninos y masculinos en los guiones de películas de Hollywood.

En el sitio oficial, puede verse a traves de una serie de visualziaciones interactivas y muy intuitivas, la participación en cantidad de diálogos entre actores hombres y mujeres para más de 2.000 guiónes de peliculas famosas.

Extraído del sitio oficial del proyecto:

>Lately, Hollywood has been taking so much shit for rampant sexism and racism. The prevailing theme: white men dominate movie roles. **But it’s all rhetoric and no data**, which gets us nowhere in terms of having an informed discussion. How many movies are actually about men? What changes by genre, era, or box-office revenue? What circumstances generate more diversity?

Veamos algunas de estas visualizaciones para comprender un poco mejor de lo que estamos hablando. En la siguiente gráfica interactiva, se muestran una cantidad de datos bien interesantes. Lo que esta graficado es la distribución de palabras para cada pelicula, entre hombres y mujeres. En el extremo izquierdo tenemos aquellas peliculas en las que los dialogos son exclusivamente de actores masculinos, mientras que en el extremo derecho tenemos las peliculas cuyos diálogos pertenecen exclusivamente a actrices femeninas. Luego a medida que nos alejamos de los extremos podemos ver las peliculas cuya distribución de dialogos entre ambos géneros varian.

![img1](https://github.com/efviodo/data-science/blob/master/courses/utec/figures/FilmDialogueFigure1.png?raw=true)

Por otro lado, el largo de la barra en la que se representa la distribución no tiene una escala constánte, si no que contempla la cantidad de peliculas asociadas a cada distribución. De esta forma si hay más peliculas en las que dominan los diálogos en un género sobre otro, el centro del gŕafico (punto donde la distribución es 50/50) no estará en el centro del gáfico, si no que por el contrario estará más cerca de uno de los extremos.

Para empezar, notese cuan claro es a la hora de representar que tan equitativo o no, son los guinoes de las películas. Segundo, cuan claro es para condensar mucha información en una sola imagen, por ejemplo la distribución de los dialogos por género minuto a minuto, para una misma película.
